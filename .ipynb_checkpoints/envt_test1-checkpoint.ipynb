```python
# Sklearn is used for data preprocessing. I will use a combination of both
# matplotlib and sklearn to convert digits into a hand writen images.
from sklearn.datasets import load_digits
my_digits = load_digits()
my_digits.images.shape
my_digits.images[0]

# Now we import matplotlib to draw the graph which is going to hold our array of numbers
import matplotlib.pyplot as plt
fig, axes = plt.subplots(5,5, figsize=(4,4))
for i,ax in enumerate(axes.flat):
    ax.imshow(my_digits.images[i])
```
